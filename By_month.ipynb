```python
complaints = pd.read_csv("Consumer_Complaints.csv", engine="python")
complaints.head()
```

```python
complaints["Date received"].head()
```

```python
complaints.index = pd.to_datetime(complaints.pop('Date received'), format='%m/%d/%Y')
```

```python
complaints.head()
```

```python
complaints.tail()
```

```python
complaints = complaints[["Product", "Company","Company response"]]
```

```python
complaints.head()
```

```python
complaint_graph = complaints.groupby(complaints.index).count().resample("M")
```

```python
complaint_graph= complaint_graph.fillna(0)
```

```python
by_month = complaint_graph['Product'][:-1]
```

```python
by_month.plot()
plt.show()
```
